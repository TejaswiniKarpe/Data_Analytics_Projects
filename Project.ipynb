*Exploratory data analysis*

```python
#Taking a overlook at male to female buyer ratio with the help of countplot
ax=sns.countplot(x='Gender',data=df)
for value in ax.containers:
    ax.bar_label(value)
"""In Seaborn, a Python data visualization library built on top of Matplotlib,a countplot is a type ofplot that shows the counts of observations 
in each category of a categorical variable.It is essentially a bar plot where the x-axis represents the categories,
and the y-axis represents the frequency or count of each category in the dataset"""
```

```python
# plotting a bar chart for gender vs total amount to analysize the money spending power based on gender
sales_gen = df.groupby(['Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
type(sales_gen)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

We can see that the female buyers are more in number as well as they've spent the higher amount which shows their higher buying capacity 

**Age**

```python
#Now to analyse the data based on the age group we can plot the countplot of agegroup catagory(ie a catagorical data)
ax=sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#To analyze based on total amount spent by different age groups 
sales_agegrp=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_agegrp)
```

We can conclude that females belonging to age group 26-35 have highest buying potential

```python
df.columns
```

```python
#To know which states has contribued what part in making sales, we can plot the bar chart of number of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
#to see what state has contributed the most in terms of total spendings
state_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=state_amount)
```

```python
# we can also check which industry is contributing most in generating the revenue
# we can visualise this my plotting a countplot for the catagory of 'Product_Catagory'

sns.set(rc={'figure.figsize':(20,7)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python


sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

**Conclusion: Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category**
